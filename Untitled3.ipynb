## Data cleaning

### Subtask:
Handle potential outliers in the `total_rooms`, `total_bedrooms`, `population`, `households`, and `median_income` columns.

### Instructions
1.  For each of the identified columns (`total_rooms`, `total_bedrooms`, `population`, `households`, and `median_income`),  investigate the distribution further using box plots and histograms.  This will help visualize the extent of the outliers and inform the cleaning strategy.

2.  Decide on an appropriate outlier handling method for each column. Options include:
    * **Winsorizing:** Cap the extreme values at a certain percentile (e.g., 95th or 99th).
    * **Trimming:** Remove data points beyond a certain percentile.
    * **Transformation:** Apply a logarithmic or other transformation to reduce the impact of outliers.
    * **No action:** If the outliers are deemed to be legitimate data points, no action is necessary.  Justify this decision in your findings.

3. Implement the chosen method for each column.  Create new columns with the cleaned data, preserving the original columns for comparison. For example, create `total_rooms_cleaned`, `total_bedrooms_cleaned`, etc.

4.  After cleaning, re-examine the descriptive statistics and visualizations (box plots and histograms) for the cleaned columns to verify the effectiveness of the outlier handling.  Document the chosen outlier handling method and its rationale for each column in the results.


**Reasoning**:
Visualize the distribution of the identified columns using box plots and histograms to understand the extent of outliers.



```python
import matplotlib.pyplot as plt
import seaborn as sns

columns_to_check = ['total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']

for col in columns_to_check:
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')

    plt.subplot(1, 2, 2)
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')

    plt.tight_layout()
    plt.show()
```

**Reasoning**:
Based on the visualizations, decide on an outlier handling method for each column and implement it, creating new cleaned columns.  Then, re-examine the distributions.



```python
import numpy as np

# Winsorize the specified columns at the 99th percentile
for col in ['total_rooms', 'total_bedrooms', 'population', 'households']:
    upper_limit = np.percentile(df[col], 99)
    df[f'{col}_cleaned'] = np.where(df[col] > upper_limit, upper_limit, df[col])

# Log transform 'median_income' to address right skew and reduce outlier influence
df['median_income_cleaned'] = np.log1p(df['median_income'])

# Re-examine the distributions after cleaning
for col in ['total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']:
    cleaned_col = f'{col}_cleaned' if col != 'median_income' else cleaned_col
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.boxplot(x=df[cleaned_col])
    plt.title(f'Box Plot of {cleaned_col}')
    plt.subplot(1, 2, 2)
    sns.histplot(df[cleaned_col], kde=True)
    plt.title(f'Histogram of {cleaned_col}')
    plt.tight_layout()
    plt.show()
```

## Data analysis

### Subtask:
Analyze the cleaned data to understand the relationships between features and the target variable.


**Reasoning**:
Calculate the correlation matrix and create scatter plots to visualize the relationship between features and the target variable.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
numerical_features = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value', 'total_rooms_cleaned', 'total_bedrooms_cleaned', 'population_cleaned', 'households_cleaned', 'median_income_cleaned']
correlation_matrix = df[numerical_features].corr()

# Display the correlation matrix
display(correlation_matrix)

# Focus on correlations with 'median_house_value'
correlations_with_target = correlation_matrix['median_house_value'].sort_values(ascending=False)
print(correlations_with_target)


# Create scatter plots for the most correlated features
most_correlated_features = correlations_with_target[1:4].index  # Exclude 'median_house_value' itself

plt.figure(figsize=(15, 5 * len(most_correlated_features)))
for i, feature in enumerate(most_correlated_features):
    plt.subplot(len(most_correlated_features), 2, 2 * i + 1)
    sns.scatterplot(x=df[feature], y=df['median_house_value'])
    plt.title(f'Scatter plot of {feature} vs median_house_value')

    cleaned_feature = feature + "_cleaned"
    if cleaned_feature in df.columns:
        plt.subplot(len(most_correlated_features), 2, 2 * i + 2)
        sns.scatterplot(x=df[cleaned_feature], y=df['median_house_value'])
        plt.title(f'Scatter plot of {cleaned_feature} vs median_house_value')

plt.tight_layout()
plt.show()
```

## Data visualization

Visualize the relationships between features and the target variable 'median_house_value' using appropriate plots.


**Reasoning**:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical features
plt.figure(figsize=(15, 10))
for i, col in enumerate(df.select_dtypes(include=['float64', 'int64']).columns):
    plt.subplot(4, 4, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()


# Scatter plots for important features vs median_house_value
important_features = ['median_income_cleaned', 'total_rooms_cleaned', 'housing_median_age']
plt.figure(figsize=(15, 10))
for i, feature in enumerate(important_features):
    plt.subplot(1, len(important_features), i + 1)
    sns.scatterplot(x=df[feature], y=df['median_house_value'])
    plt.title(f'Scatter plot of {feature} vs median_house_value')
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
